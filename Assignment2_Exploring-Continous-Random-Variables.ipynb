```python
# (a) Continous Uniform

def plotting(x, y, s):
    print("\n")
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, color="g", linewidth=2.0)
    plt.xlabel('x-axis')
    plt.ylabel('Probability')
    plt.title(f'{s} of Continuous Uniform Distribution')
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.show()

def continuous_uniform_pdf(a, b, n):
    x = np.linspace(a, b, n)
    y = [1/(b - a)] * len(x)
    plotting(x, y, "PDF")

def continuous_uniform_cdf(a, b, n):
    x = np.linspace(a, b, n)
    y = (x - a) / (b - a)
    plotting(x, y, "CDF")

a = float(input("Enter the lower bound (a): "))
b = float(input("Enter the upper bound (b): "))
n = int(b-a)*2 #arbitarily taken

continuous_uniform_pdf(a, b, n)
continuous_uniform_cdf(a, b, n)
```

```python
# (b) Gaussian Distribution

def plotting(x, y, s):
    print("\n")
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, color="b", linewidth=2.0)
    plt.fill_between(x,y, alpha=0.3, color='g')
    plt.xlabel('x-axis')
    plt.ylabel('Probability')
    plt.title(f'{s} of Gaussian Distribution')
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.show()

def gaussian_pdf(x, mu, sigma):
    return (1.0 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-1 * (1.0 / 2.0) * ((x - mu) / sigma)**2)

def continuous_gaussian_pdf(n, mu, sigma):
    x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, n) #adjustment to accomodate the mean in the centre
    y = gaussian_pdf(x, mu, sigma)
    plotting(x, y, "PDF")

def continuous_gaussian_cdf(n, mu, sigma):
    x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, n)
    y = gaussian_pdf(x, mu, sigma)
    y = np.cumsum(gaussian_pdf(x, mu, sigma)) * 1/np.sum(y)
    #if we do not use 1/np.sum(y), then we should use the correct function for calculating CDF of Gaussian,instead of Gaussian
    # -> this involves error function
    plotting(x, y, "CDF")

print("Enter mu(mean) and sigma(std deviation) (space-separated):")
mu, sigma = (float(i) for i in input().split(" "))
n = 10000

continuous_gaussian_pdf(n, mu, sigma)
continuous_gaussian_cdf(n, mu, sigma)
```

```python
# (c) Exponential Distribution

def plotting(x, y, s):
    print("\n")
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, color="r", linewidth=2.0)
    plt.fill_between(x,y, alpha=0.3, color='g')
    plt.xlabel('x-axis')
    plt.ylabel('Probability')
    plt.title(f'{s} of Exponential Distribution')
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.show()

def exponential_pdf(x, alpha):
    return alpha * np.exp(-alpha * x)

def continuous_exp_pdf(n, alpha):
    x = np.linspace(0, 5/alpha, n*n) #here i am normalising with 5 times the mean of exponential distribution
    y = exponential_pdf(x, alpha=alpha)
    plotting(x, y, "PDF")

def continuous_exp_cdf(n, alpha):
    x = np.linspace(0, 5/alpha, n*n)
    y = 1 - np.exp(-alpha * x)  #CDF
    plotting(x, y, "CDF")

print("enter lambda: ")
ld = float(input())  # ld = lambda
n = 100

continuous_exp_pdf(n, ld)
continuous_exp_cdf(n, ld)
```

```python
# (d) Cauchy Distribution


def plotting(x, y, s):
    print("\n")
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, color="b", linewidth=2.0)
    plt.fill_between(x,y, alpha=0.3, color='g')
    plt.xlabel('x-axis')
    plt.ylabel('Probability')
    plt.title(f'{s} of Cauchy Distribution')
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.show()

def cauchy_fn(x_, x0, gamma):
    return 1.0 / (np.pi * gamma * (1 + ((x_ - x0) / gamma)**2))

def continuous_cauchy_pdf(x0, gamma, n):
    x = np.linspace(x0 - 5 * gamma, x0 + 5 * gamma, n*n)  # x0 is median, gamma is scale parameter, so adjustment
    y = cauchy_fn(x,x0,gamma)
    plotting(x, y, "PDF")

def continuous_cauchy_cdf(x0, gamma, n):
    x = np.linspace(x0 - 5 * gamma, x0 + 5 * gamma, n*n)
    y = np.arctan((x - x0) / gamma) / np.pi + 0.5
    plotting(x, y, "CDF")


x0 = float(input("Enter x0 - location parameter"))
gamma = float(input("Enter gamma - scale parameter"))
n = 100

continuous_cauchy_pdf(x0, gamma, n)
continuous_cauchy_cdf(x0, gamma, n)
```

```python
# (e) Laplacian Distribution

def plotting(x, y, s):
    print("\n")
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, color="r", linewidth=2.0)
    plt.fill_between(x,y, alpha=0.3, color='g')
    plt.xlabel('x-axis')
    plt.ylabel('Probability')
    plt.title(f'{s} of Laplace Distribution')
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.show()

def laplace_pdf(x, mu, b):
    return 1.0 / (2 * b) * np.exp(-np.abs(x - mu) / b)

def continuous_laplace_pdf(mu, b, n):
    x = np.linspace(mu - 4 * b, mu + 4 * b, n*n)  #adjustment such that mean is in the centre of the graph
    y = laplace_pdf(x, mu, b)
    plotting(x, y, "PDF")


def laplace_cdf(var, mu, b):
    if var < mu:
      cdf_value = 0.5 * np.exp((var - mu) / b)
    else:
      cdf_value = 1 - 0.5 * np.exp(-(var - mu) / b)

    return cdf_value

def continuous_laplace_cdf(mu, b, n):
    x = np.linspace(mu - 4 * b, mu + 4 * b, n*n)
    y = []
    for v in x:
      y.append(laplace_cdf(v,mu,b))
    plotting(x, y, "CDF")

mu = float(input("enter mu - location parameter "))
b = float(input("enter b - scale "))

n = 100

continuous_laplace_pdf(mu, b, n)
continuous_laplace_cdf(mu, b, n)
```

```python
# (f) Rayleigh Distribution

def plotting(x, y, s):
    print("\n")
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, color="b", linewidth=2.0)
    plt.fill_between(x,y, alpha=0.3, color='g')
    plt.xlabel('x-axis')
    plt.ylabel('Probability')
    plt.title(f'{s} of Rayleigh Distribution')
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.show()

def rayleigh_pdf(x, sigma):
    return x / (sigma**2) * np.exp(-x**2 / (2 * sigma**2))

def continuous_rayleigh_pdf(sigma, n):
    x = np.linspace(0, 4 * sigma, n*n)
    y = rayleigh_pdf(x, sigma=sigma)
    plotting(x, y, "PDF")

def continuous_rayleigh_cdf(sigma, n):
    x = np.linspace(0, 4 * sigma, n*n)
    y = 1 - np.exp(-x**2 / (2 * sigma**2))
    plotting(x, y, "CDF")

print("enter the value of Sigma ")
sigma = float(input())
n = 100
continuous_rayleigh_pdf(sigma, n)
continuous_rayleigh_cdf(sigma, n)
```

```python
# (a) Discrete Uniform

def plotting(x,y):
  plt.figure(figsize=(6, 4))
  plt.bar(x,y, width=0.5, alpha=0.8, label='Discrete Uniform', color='green')
  plt.xlabel('x-axis')
  plt.ylabel('Probability')
  plt.title('CDF of Discrete Uniform Distribution')
  plt.grid(axis='y', linestyle='--',alpha=0.4)
  plt.legend()
  plt.show()


def discrete_uniform(n):
  x=np.linspace(0,n,n)
  sum = 0
  y=[]
  for k in range(1,n+1,1):
    y.append(sum)
    sum+=1/n
  plotting(x,y)


n=int(input())
discrete_uniform(n)


```

```python
# (b) Bernoulli's Distribuion

pro_of_success = [0.3]

def plotting(p,q):
    plt.figure(figsize=(6, 4))
    plt.bar(0, height=p, align='center', alpha=0.7, color="red", edgecolor="black")
    plt.bar(1, height=q, align='center', alpha=0.7, color="blue", edgecolor="black")
    plt.xlabel('Success vs Failure')
    plt.ylabel('Probability')
    plt.title('CDF of Bernoulli Distribution')
    plt.grid(axis='y', linestyle='--',alpha=0.4)
    plt.ylim(0, 1)
    plt.xticks([])
    plt.legend()
    plt.show()

def bernoulli(n):
    p=n
    q= (1-n) + (p)
    plotting(round(p,2),round(q,2))

for prob in pro_of_success:
  bernoulli(prob)


```

```python
# (c) Binomial Distribuion

bino_var = [[5, 0.7]] #nested lists with n, p in them

def plotting(y, q):
    x_values = np.arange(len(y)) # linearly incremeneting x axis values
    plt.figure(figsize=(6, 4))
    plt.bar(x_values, height=y, align='center', alpha=0.7, label='Binomial Distribution', color="green", edgecolor="red")
    plt.xlabel('x-axis')
    plt.ylabel('Probability')
    plt.title(f'CDF of Binomial Distribution for {1-q} rate of success')
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.legend()
    plt.show()

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def nCr(n, k):
    if k==0:
      return n
    else:
      return int(factorial(n) // (factorial(k) * factorial(n - k)))

def binomial(l):
    n = l[0]
    p = l[1]
    q = 1 - p
    y = []
    sum=0

    for k in range(n + 1):
        ncr = nCr(n, k)
        term = float(ncr) * (p**k) * (q**(n - k))  # Ensure term is treated as float
        sum+=term
        y.append(sum)

    plotting(y, q)

for bino in bino_var:
  binomial(bino)

```

```python
# (d) Poisson Distribution

M = 10 # number of events
probability = 0.8 # assume
lambda_ = M *probability

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def poisson_pmf(k, l):
    return (np.exp(-l) * l**k) / factorial(k)

def poisson_distribution(l,M):
    k_values = np.arange(0, M+1)
    pmf_values = [poisson_pmf(k, l) for k in k_values]
    y=[]
    sum=0
    for p in pmf_values:
      sum+=p
      y.append(sum)
    plotting(y, k_values, l)

def plotting(p, k_values, l):
    plt.figure(figsize=(6, 4))
    plt.bar(k_values,width = 0.5, height=p, align='center', alpha=0.65, color='blue', edgecolor='red')
    plt.xlabel('Number of Events')
    plt.ylabel('Probability')
    plt.title(f'Poisson Distribution (λ={l})')
    plt.show()

poisson_distribution(lambda_,M)


```

```python
# (e) Geometric Distribution

values = [0.01]
def geometric_pmf(k, p):
    return ((1 - p) ** (k - 1)) * p

def plotting(p, k_values, prob_of_success):
    plt.figure(figsize=(6, 4))
    plt.bar(k_values, height=p,width=0.5, align='center', alpha=0.65, color='black')
    plt.xlabel('Number of Trials Until First Success')
    plt.ylabel('Probability')
    plt.title(f'Geometric Distribution (p={prob_of_success})')
    plt.show()


def geometric(v):
  k_values_geometric = np.arange(1, 101)  # geometric distribution starts from 1 not 0. 1 to 101 picked arbitarily.
  pmf_values_geometric = [geometric_pmf(k, v) for k in k_values_geometric]
  y=[]
  sum=0
  for p in pmf_values_geometric:
    sum+=p
    y.append(sum)
  plotting(y, k_values_geometric, v)


for v in values:
    geometric(v)

```
